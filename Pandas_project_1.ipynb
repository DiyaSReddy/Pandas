```python
import numpy as np
import pandas as pd

# Load dataset and clean missing values
df = pd.read_csv("/content/customer.csv")
df.fillna({'Returns': 0, 'Total Purchase Amount': 0}, inplace=True)

# 2. Analyze cutsomer purchase behaviour
most_purchased = df['Product Category'].value_counts().head(5)
high_spenders = df.groupby('Customer ID')['Total Purchase Amount'].sum().sort_values(ascending=False).head(5)
print("\nTop 5 Most Purchased Categories:\n", most_purchased)
print("\nTop 5 Highest Spending Customers:\n", high_spenders)

# 3. Average order value and top categories
total_revenue = df['Total Purchase Amount'].sum()
total_orders = df.shape[0]  # Total transactions
average_order_value = total_revenue / total_orders
print("\nAverage Order Value (AOV): ",average_order_value)

top_selling = df.groupby('Product Category')['Total Purchase Amount'].sum().sort_values(ascending=False).head(5)
print("\nTop 5 Categories: \n", top_selling)

# 4️. Seasonal Trends in Purchases
df['Purchase Date'] = pd.to_datetime(df['Purchase Date']) # Convert 'Purchase Date' to datetime
df['Month'] = df['Purchase Date'].dt.month
df['Year'] = df['Purchase Date'].dt.year
monthly_trends = df.groupby(['Year', 'Month'])['Total Purchase Amount'].sum()
print("\nMonthly Trends in purchases:\n", monthly_trends.tail(24))
```
