```python
import pandas as pd
import numpy as np

# Load the dataset from uploaded file
df = pd.read_csv("/content/netflix.csv")

#1. Count of Movies vs. TV Shows
count = df['type'].value_counts()
print("\nMovies vs TV Shows Count:\n", count)

#2. Displaying top 5 most Popular Genres
#splitting based on commas and making new array as one title can be in multiple genres
genre = np.concatenate(df['listed_in'].str.split(', ').values)
genre_count = pd.Series(genre).value_counts().head(5)
print("\nTop 5 Genres:\n", genre_count)

# 3. Release Trends Over the Years
df['release_year'] = pd.to_numeric(df['release_year'])
yearcount = df['release_year'].value_counts().sort_index()
print("\nRelease Trends Over the Years:\n", yearcount.tail(5))

# 4. Top Directors with the Most Content
top_directors = df['director'].value_counts().head(5)
print("\nTop 5 Directors with Most Content:\n", top_directors)
```
